```python
def calculate_housing_needs(population, FApC,
                                population_scenario=population.scenario, FApC_scenario=FApC.scenario,
                                FApC_variable=None,
                                model_name=None,
                                scenario_name=None,
                                variable_name=None,
                                unit_name=None):

        '''
        Calculate the housing needs based on the population and the m2/capita scenarios as variables
        Returns a IamDataframe with the housing needs in m2
        '''

        # Filter population by scenario if specified and transform iamdf to df in order to perform calculations
        if population_scenario:
            population = population.filter(scenario=population_scenario)
            population_df = population.as_pandas()

        # Filter FApC by scenario and variable if specified and transform iamdf to df in order to perform calculations
        if FApC_scenario:
            FApC = FApC.filter(scenario=FApC_scenario, variable=FApC_variable)
            FApC_df = FApC.as_pandas()

        # Multiply population and FApC timeseries to get housing needs
        housing_needs = population_df['value'] * FApC_df['value']

        # create a new dataframe with year as a column
        result_df = pd.DataFrame({'year': population_df['year'],
                                  'value': housing_needs})

        # Create a new IamDataFrame object from the housing needs timeseries
        housing_needs_iamdf = pyam.IamDataFrame(result_df,
                                         model=model_name,
                                         scenario=scenario_name,
                                         region='France',
                                         variable=variable_name,
                                         unit=unit_name)

        return housing_needs_iamdf
```

```python
def calculate_all_housing_needs(population, FApC,
                            population_scenarios=None, FApC_scenarios=None,
                            FApC_variable=None,
                            unit_name=None):

    '''
    Calculate the housing needs based on the population and the m2/capita scenarios as variables
    Returns an IamDataFrame with the housing needs in m2
    '''

    housing_needs_list = []  # List to store the calculated housing needs for each combination

    for pop_scenario in population_scenarios:
        population_filtered = population.filter(scenario=pop_scenario)
        population_df = population_filtered.as_pandas()

        for FApC_scenario in FApC_scenarios:
            FApC_filtered = FApC.filter(scenario=FApC_scenario, variable=FApC_variable)
            FApC_df = FApC_filtered.as_pandas()

            # Multiply population and FApC timeseries to get housing needs
            housing_needs = population_df['value'] * FApC_df['value']

            # Create a new dataframe with year as a column
            result_df = pd.DataFrame({'year': population_df['year'],
                                      'value': housing_needs})

            # Get the model and scenario names from population and FApC scenarios
            model_name = population_filtered['model'].iloc[0]
            scenario_name = population_filtered['scenario'].iloc[0] + '_' + FApC_filtered['scenario'].iloc[0]
            
            # Create a new IamDataFrame object from the housing needs timeseries
            housing_needs_iamdf = pyam.IamDataFrame(result_df,
                                             model=model_name,
                                             scenario=scenario_name,
                                             region='France',
                                             variable='Housing needs',
                                             unit=unit_name)

            housing_needs_list.append(housing_needs_iamdf)

    # Concatenate all the housing needs IamDataframes into a single IamDataFrame
    combined_housing_needs_iamdf = pyam.concat(housing_needs_list)

    return combined_housing_needs_iamdf
```

```python
def plot_scenarios(iamdf, title=None):
    """
    Plot housing needs from the given IamDataFrame and display the legend outside the plot.
    """
    fig, ax = plt.subplots(figsize=(12, 8))

    # Plot the housing needs with color based on scenario
    iamdf.plot(ax=ax, color="scenario", legend=False)

    # Retrieve the legend handles and labels
    handles, labels = ax.get_legend_handles_labels()

    # Get the values for sorting the legend
    values = [handle.get_ydata()[-1] for handle in handles]

    # Sort the legend handles, labels, and values in descending order
    handles, labels, values = zip(*sorted(zip(handles, labels, values), key=lambda x: x[2], reverse=True))

    # Create a new legend with the sorted handles and labels outside the plot
    ax.legend(handles, labels, bbox_to_anchor=(1.02, 1), loc='upper left')

    # Adjust the plot to accommodate the legend outside
    plt.subplots_adjust(right=0.75)

    # Add a title to the plot
    ax.set_title(title)

    # Show the plot
    plt.show()
```

### Construction from population growth

```python
def calculate_construction_needs_from_population_growth(housing_needs_iamdf):
        """
        Calculates the construction needs from population growth based on housing needs data.
        """

        # calculate the difference in housing needs from t to t-1
        construction_needs = housing_needs_iamdf.as_pandas()['value'].diff().fillna(0)

        # replace negative values with 0
        construction_needs = construction_needs.clip(lower=0)

        # create a new dataframe with year as a column
        result_df = pd.DataFrame({'year': housing_needs_iamdf['year'],
                                  'value': construction_needs})

        # create a new IamDataFrame object from the construction needs data
        construction_needs_from_population_growth_iamdf = pyam.IamDataFrame(result_df,
                                                     model='FF_INSEE',
                                                     scenario='BAU',
                                                     region='France',
                                                     variable='Construction needs from population growth',
                                                     unit='m2')

        return construction_needs_from_population_growth_iamdf
```

```python
def calculate_construction_needs_from_population_growth_combined(housing_needs_combined_iamdf):

    '''
    Calculate the housing needs based on the population and the m2/capita scenarios as variables
    Returns an IamDataFrame with the housing needs in m2
    '''

    construction_needs_list = []  # List to store the calculated housing needs for each combination
    
    for scenarios in housing_needs_combined_iamdf.scenario:
        housing_needs_filtered = housing_needs_combined_iamdf.filter(scenario=scenarios)
        housing_needs_df = housing_needs_filtered.as_pandas()
        
        # calculate the difference in housing needs from t to t-1
        construction_needs = housing_needs_df['value'].diff().fillna(0)

        # replace negative values with 0
        construction_needs = construction_needs.clip(lower=0)

        # create a new dataframe with year as a column
        result_df = pd.DataFrame({'year': housing_needs_combined_iamdf['year'],
                              'value': construction_needs})
        
        # Get the model and scenario names from housing_needs
        model_name = housing_needs_filtered['model'].iloc[0]
        scenario_name = housing_needs_filtered['scenario'].iloc[0] 
            
        # Create a new IamDataFrame object from the housing needs timeseries
        construction_needs_iamdf = pyam.IamDataFrame(result_df,
                                             model=model_name,
                                             scenario=scenario_name,
                                             region='France',
                                             variable='Construction needs from population growth',
                                             unit='m2')

        construction_needs_list.append(construction_needs_iamdf)

        # Concatenate all the housing needs IamDataframes into a single IamDataFrame
        combined_construction_needs_from_population_growth_iamdf = pyam.concat(construction_needs_list)

    return combined_construction_needs_from_population_growth_iamdf
```

```python
class StockModel:
    def __init__(self, stock, vacant_stock):
        self.stock = stock
        self.vacant_stock = vacant_stock

    def run_simulation(self, population, dem, vacancy, construction_needs_scenario=None):
        
        # Create empty lists to store results
        years = []
        demolition_surface_results = []
        vacant_surface_available_results = []
        effective_construction_results = []
        stock_results = []
        vacant_stock_results = []

        # Iterate over years
        for year in range(population['year'].min() + 1, population['year'].max() + 1):
             
            # Calculate demolition surface
            stock_demolition = self.stock  # Previous year's stock
            demolition_surface = stock_demolition * dem.filter(year=year, variable='Dwellings').data['value'].values[0]

            # Calculate vacant surface
            stock_vacant = self.vacant_stock  # Previous year's vacant stock
            vacant_surface_available = stock_vacant * vacancy.filter(year=year, variable='%vacancy_construction').data['value'].values[0]

            # Calculate construction needs 
            construction_needs = demolition_surface + combined_construction_needs_from_population_growth_iamdf.filter(year=year, scenario=construction_needs_scenario).data['value'].values[0]
      
            # Calculate effective construction needs, taking into account the vacant surface that becomes available to meet construction needs
            effective_construction = construction_needs - vacant_surface_available
        
            # Update stock and vacant stock
            self.stock += effective_construction
            self.vacant_stock -= vacant_surface_available

            # Append results to lists
            years.append(year)
            demolition_surface_results.append(demolition_surface)
            vacant_surface_available_results.append(vacant_surface_available)
            effective_construction_results.append(effective_construction)
            stock_results.append(self.stock)
            vacant_stock_results.append(self.vacant_stock)

        # Create dataframes from results
        results_df = pd.DataFrame({
            'Years': years,
            'Demolition_surface': demolition_surface_results,
            'Vacant_surface_available': vacant_surface_available_results,
            'Effective_construction': effective_construction_results,
            'Stock': stock_results,  # Final stock
            'Vacant_stock': vacant_stock_results  
        })

        return results_df
```

```python
class StockModel:
    def __init__(self, stock, vacant_stock, stock_ind, stock_col, stock_dwelling, stock_dwelling_ind, stock_dwelling_col):
        self.stock = stock
        self.vacant_stock = vacant_stock
        self.stock_ind = stock_ind
        self.stock_col = stock_col
        self.stock_dwelling = stock_dwelling
        self.stock_dwelling_ind = stock_dwelling_ind
        self.stock_dwelling_col = stock_dwelling_col 

    def run_simulation(self, dem_scenario=None, vacancy_scenario=None, construction_needs_scenario=None,
                        pct_ind_col_surface_scenario=None, new_scenario=None):
        
        # Create empty lists to store results
        years = []
        demolition_surface_results = []
        vacant_surface_available_results = []
        effective_construction_results = []
        effective_construction_ind_results = []
        effective_construction_col_results = []
        new_dwelling_ind_results = []
        new_dwelling_col_results = []
        new_dwelling_results = []
        stock_results = []
        stock_ind_results = []
        stock_col_results = []
        vacant_stock_results = []
        stock_dwelling_results = []
        stock_dwelling_ind_results = []
        stock_dwelling_col_results = []

        # Iterate over years
        for year in range(2020 + 1, 2050 + 1):
            
            # Calculate demolition surface
            stock_demolition = self.stock  # Previous year's stock
            demolition_surface = stock_demolition * dem.filter(year=year, scenario=dem_scenario, variable='Dwellings').data['value'].values[0]

            # Calculate vacant surface
            stock_vacant = self.vacant_stock  # Previous year's vacant stock
            vacant_surface_available = stock_vacant * vacancy.filter(year=year, scenario=vacancy_scenario, variable='%vacancy_construction').data['value'].values[0]

            # Calculate construction needs 
            construction_needs = demolition_surface + combined_construction_needs_from_population_growth_iamdf.filter(year=year, scenario=construction_needs_scenario).data['value'].values[0]
      
            # Calculate effective construction needs, taking into account the vacant surface that becomes available to meet construction needs
            effective_construction = construction_needs - vacant_surface_available
        
            # Update stock and vacant stock
            self.stock += effective_construction
            self.vacant_stock -= vacant_surface_available

            # Calculate individual and collective construction surface based on the effective construction
            effective_construction_ind = effective_construction * pct_ind_col_surface.filter(year=year, scenario=pct_ind_col_surface_scenario, variable='pct_ind_new_surface').data['value'].values[0]
            effective_construction_col = effective_construction * pct_ind_col_surface.filter(year=year, scenario=pct_ind_col_surface_scenario, variable='pct_col_new_surface').data['value'].values[0]
            
            # Convert individual and collective construction surface to number of dwellings
            new_dwelling_ind = effective_construction_ind / new.filter(year=year, scenario=new_scenario, variable='new_ind_surface_average').data['value'].values[0]
            new_dwelling_col = effective_construction_col / new.filter(year=year, scenario=new_scenario, variable='new_col_surface_average').data['value'].values[0]
            new_dwelling = new_dwelling_ind + new_dwelling_col

            # Update stock_ind and stock_col (surface)
            self.stock_ind += effective_construction_ind
            self.stock_col += effective_construction_col
            
            # Update stock_dwelling_ind and stock_dwelling_col 
            self.stock_dwelling_ind += new_dwelling_ind
            self.stock_dwelling_col += new_dwelling_col
            self.stock_dwelling = self.stock_dwelling_ind + self.stock_dwelling_col
    
            # Append results to lists
            years.append(year)
            demolition_surface_results.append(demolition_surface)
            vacant_surface_available_results.append(vacant_surface_available)
            effective_construction_results.append(effective_construction)
            effective_construction_ind_results.append(effective_construction_ind)
            effective_construction_col_results.append(effective_construction_col)
            new_dwelling_ind_results.append(new_dwelling_ind)
            new_dwelling_col_results.append(new_dwelling_col)
            new_dwelling_results.append(new_dwelling)
            stock_results.append(self.stock)
            stock_ind_results.append(self.stock_ind)
            stock_col_results.append(self.stock_col)
            vacant_stock_results.append(self.vacant_stock)
            stock_dwelling_results.append(self.stock_dwelling)
            stock_dwelling_ind_results.append(self.stock_dwelling_ind)
            stock_dwelling_col_results.append(self.stock_dwelling_col)
        
        
        results_df = pd.DataFrame({
            'Years': years,
            'Demolition_surface': demolition_surface_results,
            'Vacant_surface_available': vacant_surface_available_results,
            'Effective_construction': effective_construction_results,
            'Effective_construction_ind': effective_construction_ind_results,
            'Effective_construction_col': effective_construction_col_results,
            'New_dwelling_ind': new_dwelling_ind_results,
            'New_dwelling_col':new_dwelling_col_results,
            'New_dwelling': new_dwelling_results,
            'Stock': stock_results,  # Final stock
            'Stock_ind': stock_ind_results,
            'Stock_col': stock_col_results,
            'Stock_dwelling': stock_dwelling_results,
            'Stock_dwelling_ind': stock_dwelling_ind_results,
            'Stock_dwelling_col': stock_dwelling_col_results,
            'Vacant_stock': vacant_stock_results,
        })
        
        return results_df
    
    def archetypes_dwellings(self, stock_model_df, pct_new_archetype, archetype_scenario=None):
        
        # Create empty lists to store results
        years = []
        new_dwelling_ind_results = []
        new_dwelling_col_results = []
        new_dwelling_results = []
        new_dwelling_ind_arch_A_results = []
        new_dwelling_ind_arch_B_results = []
        new_dwelling_ind_arch_C_results = []
        new_dwelling_ind_arch_D_results = []
        new_dwelling_col_arch_E_results = []
        new_dwelling_col_arch_F_results = []
        new_dwelling_col_arch_G_results = []
        new_dwelling_col_arch_H_results = []

        # Iterate over years
        for year in range(2020 + 1, 2050 + 1):
    
            # Take individual and collective new dwellings previously calculated
            new_dwelling_ind = stock_model_df.loc[stock_model_df['Years'] == year, 'New_dwelling_ind'].iloc[0]
            new_dwelling_col = stock_model_df.loc[stock_model_df['Years'] == year, 'New_dwelling_col'].iloc[0]
            new_dwelling = stock_model_df.loc[stock_model_df['Years'] == year, 'New_dwelling'].iloc[0]

            # Calculate the number of each archetypes based on the scenario archetype distributions
            new_dwelling_ind_arch_A = new_dwelling_ind * pct_new_archetype.filter(year=year, scenario=archetype_scenario, variable='pct_archetype_ind_A').data['value'].iloc[0]
            new_dwelling_ind_arch_B = new_dwelling_ind * pct_new_archetype.filter(year=year, scenario=archetype_scenario, variable='pct_archetype_ind_B').data['value'].iloc[0]
            new_dwelling_ind_arch_C = new_dwelling_ind * pct_new_archetype.filter(year=year, scenario=archetype_scenario, variable='pct_archetype_ind_C').data['value'].iloc[0]
            new_dwelling_ind_arch_D = new_dwelling_ind * pct_new_archetype.filter(year=year, scenario=archetype_scenario, variable='pct_archetype_ind_D').data['value'].iloc[0]
            new_dwelling_col_arch_E = new_dwelling_col * pct_new_archetype.filter(year=year, scenario=archetype_scenario, variable='pct_archetype_col_E').data['value'].iloc[0]
            new_dwelling_col_arch_F = new_dwelling_col * pct_new_archetype.filter(year=year, scenario=archetype_scenario, variable='pct_archetype_col_F').data['value'].iloc[0]
            new_dwelling_col_arch_G = new_dwelling_col * pct_new_archetype.filter(year=year, scenario=archetype_scenario, variable='pct_archetype_col_G').data['value'].iloc[0]
            new_dwelling_col_arch_H = new_dwelling_col * pct_new_archetype.filter(year=year, scenario=archetype_scenario, variable='pct_archetype_col_H').data['value'].iloc[0]

            # Append results to list 
            years.append(year)
            new_dwelling_ind_results.append(new_dwelling_ind)
            new_dwelling_col_results.append(new_dwelling_col)
            new_dwelling_results.append(new_dwelling)
            new_dwelling_ind_arch_A_results.append(new_dwelling_ind_arch_A)
            new_dwelling_ind_arch_B_results.append(new_dwelling_ind_arch_B)
            new_dwelling_ind_arch_C_results.append(new_dwelling_ind_arch_C)
            new_dwelling_ind_arch_D_results.append(new_dwelling_ind_arch_D)
            new_dwelling_col_arch_E_results.append(new_dwelling_col_arch_E)
            new_dwelling_col_arch_F_results.append(new_dwelling_col_arch_F)
            new_dwelling_col_arch_G_results.append(new_dwelling_col_arch_G)
            new_dwelling_col_arch_H_results.append(new_dwelling_col_arch_H)
            
        archetypes_df = pd.DataFrame({
            'Years': years,
            'New_dwelling_ind': new_dwelling_ind_results,
            'New_dwelling_col': new_dwelling_col_results,
            'New_dwelling': new_dwelling_results,
            'New_dwelling_ind_arch_A': new_dwelling_ind_arch_A_results,
            'New_dwelling_ind_arch_B': new_dwelling_ind_arch_B_results,
            'New_dwelling_ind_arch_C': new_dwelling_ind_arch_C_results,
            'New_dwelling_ind_arch_D': new_dwelling_ind_arch_D_results,
            'New_dwelling_col_arch_E': new_dwelling_col_arch_E_results,
            'New_dwelling_col_arch_F': new_dwelling_col_arch_F_results,
            'New_dwelling_col_arch_G': new_dwelling_col_arch_G_results,
            'New_dwelling_col_arch_H': new_dwelling_col_arch_H_results

        })
        
        return archetypes_df
```
