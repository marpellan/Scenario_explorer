```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np 
```

```python
def plot_images_as_subplots(image_paths, titles=None, num_rows=2, num_cols=2, figsize=(10, 8), hspace=None, wspace=None):
    """
    Display multiple PNG images as subplots in a single figure.

    Args:
    - image_paths (list of str): List of file paths to PNG images.
    - titles (list of str): Optional list of titles for each subplot.
    - num_rows (int): Number of rows for the subplot grid.
    - num_cols (int): Number of columns for the subplot grid.
    - figsize (tuple): Figure size (width, height) in inches.

    Returns:
    - None (displays the figure with subplots).
    """
    if len(image_paths) != num_rows * num_cols:
        raise ValueError("Number of images does not match the subplot grid size.")

    # Create a figure with subplots
    fig, axs = plt.subplots(num_rows, num_cols, figsize=figsize)

    for i, ax in enumerate(axs.flat):
        img = mpimg.imread(image_paths[i])
        ax.imshow(img)
        if titles:
            ax.set_title(titles[i])
        ax.axis('off')

    # Add some spacing between subplots
    plt.tight_layout()
    
     # Adjust subplot spacing
    fig.subplots_adjust(hspace=hspace, wspace=wspace)

    # Show the combined graph
    plt.show()
```

```python
# List of PNG image file paths
image_paths_op = ['operational_bau.png', 'operational_optimistic.png', 'cum_operational_bau.png', 'cum_operational_optimistic.png']

# Call the function to display images as subplots
plot_images_as_subplots(image_paths_op, num_rows=2, num_cols=2, figsize=(10, 8), hspace=0.00000001, wspace=0.00000001)
```

```python
# List of PNG image file paths
image_paths_op = ['embodied_bau.png', 'embodied_optimistic.png', 'cum_embodied_bau.png', 'cum_embodied_optimistic.png']

# Call the function to display images as subplots
plot_images_as_subplots(image_paths_op, num_rows=2, num_cols=2, figsize=(10, 8), hspace=0.5, wspace=0.0001)
```
